# Scrape City Coordinates

```python
import urllib.request
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_cities_in_the_Philippines'
    
req = Request(url, headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Connection': 'keep-alive'})    
    
content = urlopen(req).read()
    
soup = BeautifulSoup(content)

```

```python
myDivs = soup.find("table", {'class':"wikitable sortable"}) #cXenseParse
```

```python
cities = myDivs.findAll('a', {"class" : "external text"})
```

```python
cities[0].text
```

```python
city_coordinates = pd.DataFrame(columns = (['city', 'coordinates']))
```

```python
city_coordinates.h
```

```python
for city in cities:
    try:
        cit = re.search('(\(.*\))',city.text).group()
        coordinates = city.span.text
    except AttributeError:
        continue
    city_coordinates = city_coordinates.append(pd.Series([cit,coordinates], index = city_coordinates.columns), ignore_index = True)
```

```python
latitude = []
longitude = []
```

```python
lat, long = city_coordinates['coordinates'][0].split()
```

```python
for coor in city_coordinates['coordinates']:
    lat, long = coor.split()
    latitude.append(round(float(re.split('[°\'\′\″"]+', lat)[0]) + (float(re.split('[°\'\′\″"]+', lat)[1]) / 60 )  + (float(re.split('[°\'\′\″"]+', lat)[2]) / 3600 ), 5))
    longitude.append(round(float(re.split('[°\'\′\″"]+', long)[0]) + (float(re.split('[°\'\′\″"]+', long)[1]) / 60 )  + (float(re.split('[°\'\′\″"]+', long)[2]) / 3600 ), 5))
```

```python
city_coordinates['latitude'] = latitude
city_coordinates['longitude'] = longitude
```

```python
city_coordinates
```

```python
city_coordinates.to_csv('ph_city_coordinates.csv', index = False)
```
